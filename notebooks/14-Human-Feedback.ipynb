### Import dependencies

```python
from langsmith import Client
```

### Attach feedback to s specific trace in langsmith

```python
client = Client()
```

```python
RUN_ID="019e2015-13cf-7f62-9499-18e4b00e5241"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="user_feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a great product!",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a great product!",
    score=0,
    feedback_source_type="api"
)
```
